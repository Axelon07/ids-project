```python
import matplotlib.pyplot as plt

plt.figure()
df["duration"].hist()
plt.title("Duration Histogram")
plt.xlabel("Duration")
plt.ylabel("Count")
plt.savefig("../reports/figures/duration_hist.png", dpi=150)
plt.show()
```

```python
plt.figure()
df["label"].value_counts().plot(kind="bar")
plt.title("Label Counts (Benign vs Attack)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.savefig("../reports/figures/label_counts.png", dpi=150)
plt.show()
```
